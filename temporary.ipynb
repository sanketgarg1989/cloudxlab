```python
class DecisionTreeNode:
    def __init__(self, attribute=None, boundary=None, left=None, right=None, decision=None):
        self.attribute = attribute    # e.g., 'salary'
        self.boundary = boundary      # e.g., 1000
        self.left = left
        self.right = right
        self.decision = decision      # True/False for leaf nodes

    def check(self, features: dict) -> bool:
        # If this is a leaf node, return the final decision (True/False)
        if self.decision is not None:
            return self.decision
        

        if features[self.attribute] <= self.boundary:
            return self.left.check(features)
        else:
            return self.right.check(features)

# --- Helpers ---
def YES():
    return DecisionTreeNode(decision=True)

def NO():
    return DecisionTreeNode(decision=False)
```

```python
dt = DecisionTreeNode(
    attribute='salary', 
    boundary=1000, 
    left=NO(), 
    right=DecisionTreeNode(
        attribute='distance', 
        boundary=40, 
        left=YES(), 
        right=NO()
    )
)
```

```python
job_description = {'salary': 2000, 'distance': 1}
```

```python
print(dt.check(job_description))
```
