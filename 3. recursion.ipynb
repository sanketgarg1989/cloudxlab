```python
# Factorial using recursion

def factorial_recursive(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial_recursive(n-1)
```

```python
factorial_recursive(5)  # Output: 120  
```

```python
# Multiplication using Recursion

def multiply_recursive (a,b):
    if b == 0:
        return 0
    elif b < 0:
        return multiply_recursive(-a, -b)
    else:
        return a + multiply_recursive (a,b-1)
```

```python
multiply_recursive(7, 2) 
```

```python
# Compute power(x, n) using recursion where n is a positive integer.

def power(x, n):
    if n == 0:
        return 1
    else:
        return x * power(x, n-1)
```

```python
power(2, 3)
```

```python
# Compute power(x, n) using recursion where n is an integer i.e. it can be positive, negative or zero.

def compute_power(x, n):
    if n == 0:
        return 1
    elif n < 0:
        return x * compute_power(1/x, -n)
    else:
        return x * compute_power(x, n-1)
```

```python
compute_power(3, -3)
```

```python
# Division using recursion to find quotient and remainder

def recursive_divide(a,b):

    if b == 0:
        return "Error: Division by zero"
    elif a < b:
         return 0 , a
    else:
    
        quotient, remainder = recursive_divide(a - b, b)
        return quotient + 1 , remainder 
```

```python
recursive_divide(2, 5) 
```

```python
#Compute HCF using Euclid's Method with Recursion

def compute_hcf(a, b):
    if b == 0:
        return a
    else:
        return compute_hcf(b, a % b)
```

```python
compute_hcf(0, 5) 
```

```python
def solve_hanoi(n, source='A', auxiliary='B', target='C'):
    # Base Case: Only one disk to move
    if n == 1:
        print(f"Moving 1 from {source} to {target}")
        return 1
    
    # Recursive Case:
    # 1) Move n-1 disks from source to auxiliary (using target as temporary)
    moves_part1 = solve_hanoi(n - 1, source, target, auxiliary)
    
    # 2) Move the largest disk (n) from source to target
    print(f"Moving {n} from {source} to {target}")
    my_move = 1
    
    # 3) Move the n-1 disks from auxiliary to target (using source as temporary)
    moves_part2 = solve_hanoi(n - 1, auxiliary, source, target)
    
    # Return the total sum of moves
    return moves_part1 + my_move + moves_part2

```

```python
# Example: Solve for 3 disks
total = solve_hanoi(3)
print(f"\nTotal moves performed: {total}")
```
