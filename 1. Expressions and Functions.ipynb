```python
#Calculate Hypotense
def calculate_hypotenuse(b,h):
    return (b*b + h*h)**0.5
a=calculate_hypotenuse(3,4)
print("Hypotenuse is ",a);
```

```python
#Find Distance 2D
def find_distance_2d(x1,y1,x2,y2):
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5
a=find_distance_2d(1, 2, 4, 6)
print("Distance in 2D is ",a);
```

```python
#Find Distance 3D
def calculate_distance_3d(x1,y1,z1,x2,y2,z2):
    return ((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)**0.5
a=calculate_distance_3d(0, 0, 0, 1, 1, 1)
print("Distance in 3D is ",a);
```

```python
# manhattan_distance
def manhattan_distance(x1,y1,x2,y2):
    return abs(x2 - x1) + abs(y2 - y1)  
a=manhattan_distance(2, 3, 5, 1)
print("Manhattan Distance is ",a);
```

```python
# Predict Y on a Straight Line (2D)
def predict(m,c,x):
    return m*x + c
y=predict(2, 3, 4)  
print("Predicted Y is ",y);
```

```python
#Learn Line Equation from Two Points
def fit(x1, y1, x2, y2):
    if x2 -x1==0:
        print("slope is undefined for vertical line")
        return None,None
    else:
        m=(y2 - y1) / (x2 - x1)
        c=y1 - m*x1
        return m,c
m,c=fit(3, 2, 5, 6)
print("value of m and c are ",(m,c));
```

```python
def f(x):
    return x **2 - 10 * x +2
```

```python
f(1000)
```

```python
#Find First Derivative of a Function

#Find the rate of change of y wrt x at x = 10

x = 1000
dx = .000001
y = f(x)
y1 = f(x + dx)
rate_of_change = (y1 - y)/dx # dy/dx
rate_of_change
```

```python
def diff(ff, x):
    dx = 1e-8

    y = ff(x)
    y1 = ff(x + dx)
    rate_of_change = (y1 - y)/dx # dy/dx
    return rate_of_change
```

```python
def myfunc(x):
    return x**4

diff(myfunc, 10)
```
