```python
# Find Minimum and Maximum of a List

def find_min_max(numbers):
    data = numbers
    curr_min = data[0]
    curr_max = data[0]
    
    for i in data:
        if i < curr_min:
             curr_min = i
        elif i > curr_max:
             curr_max = i
    return curr_min,curr_max
```

```python
find_min_max([0, 7, 3, 7])  
```

```python
# Min-Max Normalization

def min_max_normalize(value, data):
    curr_min = data[0]
    curr_max = data[0]
    x = value
    for i in data:
        if i < curr_min:
             curr_min = i
        elif i > curr_max:
             curr_max = i
    norm_value = (x - curr_min)/(curr_max - curr_min)
    return norm_value
```

```python
min_max_normalize(10, [10, 20, 30])  
```

```python
# Compute Mean of a List of Numbers

def compute_mean(numbers):
    add = 0
    if not numbers:
        return 0
    for i in numbers:
        add = add + i
    #print(add)
    mean = add/len(numbers)
    return mean
```

```python
compute_mean([2, 4, 4, 4, 5, 5, 7, 9])   
```

```python
# Compute Standard Deviation (SD) of a List of Numbers

def compute_sd(numbers):
    mean = compute_mean(numbers)
    #print (mean)
    sq_diff = []
    #print(sq_diff)
    for i in numbers:
        sq_diff.append((mean - i) **2)
        #print(sq_diff)
    avg_sq_diff = compute_mean(sq_diff)
    return avg_sq_diff ** 0.5


```

```python
compute_sd([10, 10, 10, 10]) 
```

```python
# Detect Outliers with Z-Score

def find_outliers(nums, threshold):
    mean = compute_mean(nums)
    sd = compute_sd(nums)
    outliers = []

    for i in nums:
        z_score = (i - mean) / sd
        #print(z_score)
        if abs(z_score) >  threshold:
            #print(i)
            outliers.append(i)
    return outliers
```

```python
find_outliers([5, 6, 7, 8, 9, 10, 100], 2) 
```

```python
# Compute IQR

def get_median(data):
    length = len(data)
    mid = length // 2
    #print(mid)
    if length % 2 == 0:
        return (data[mid - 1] + data [mid]) / 2
    else:
        return data[mid]

def compute_iqr(data):
    data.sort()
    n = len(data)
    mid = n // 2
    if n % 2 != 0:
        lower_half = data[:mid + 1]
        upper_half = data[mid:]
    else:
        lower_half = data[:mid]
        upper_half = data[mid:]

    
    q1=get_median(lower_half)
    q3=get_median(upper_half)
    return q3 - q1

    
```

```python
compute_iqr([10, 20, 30, 40, 50, 60])  
```

```python
# Standardization
def compute_mean(numbers):
    add = 0
    if not numbers:
        return 0
    for i in numbers:
        add = add + i
    #print(add)
    mean = add/len(numbers)
    return mean

def compute_sd(numbers):
    mean = compute_mean(numbers)
    #print (mean)
    sq_diff = []
    #print(sq_diff)
    for i in numbers:
        sq_diff.append((mean - i) **2)
        #print(sq_diff)
    avg_sq_diff = compute_mean(sq_diff)
    return avg_sq_diff ** 0.5

def standardize(data):
    mean = compute_mean(data)
    sd = compute_sd(data)
    std = []
    for i in data:
        std.append((i-mean)/sd)
    return std

    
```

```python
standardize([10, 10, 10])  
```
