```python
# Check if a point P is closer to point A or point B in 2D
def closer_point(p, a, b):
    def distance(p1, p2):
        return ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
    if distance(p, a) < distance(p, b):
        return a
    else:
        return b
```

```python
closer_point((7, 3), (2, 3), (10, 3)) 
```

```python
# Are Two Lines Overlapping or Touching in 1D

def are_lines_touching_or_overlapping(start1, end1, start2, end2):
    if start1 <= start2 <= end1:
        print (True)
    elif start1 <= end2 <= end1:
        print (True)
    else:
        print (False)   
```

```python
# Are Two Lines Overlapping or Touching in 1D (Another version)

def are_lines_touching_or_overlapping(start1, end1, start2, end2):
    # Normalize segments to (min, max)
    s1, e1 = min(start1, end1), max(start1, end1)
    s2, e2 = min(start2, end2), max(start2, end2)

    # They overlap if the start of one is <= the end of the other
    if s1 > e2:
        return False
    elif s2 > e1:
        return False
    else:
        return True
```

```python
are_lines_touching_or_overlapping(-2, -7, -4, -3)
```

```python
# Is a point inside a rectangle?

def is_point_inside_rectangle(x1, y1, x2, y2, px, py):
    if min(x1,x2) <= px <= max(x1,x2) and min(y1,y2) <= py <= max(y1,y2):
        return True
    else:
        return False
    
```

```python
# Are rectangles (with sides parallel to axes) intersecting?

def are_rectangles_intersecting(rect1, rect2):
 
    (r1x1, r1y1), (r1x2, r1y2) = rect1
    (r2x1, r2y1), (r2x2, r2y2) = rect2

    if min(r1x1 ,r1x2) <= min(r2x1,r2x2) <= max(r1x1 ,r1x2) and min(r1y1 ,r1y2) <= min(r2y1,r2y2) <= max(r1y1 ,r1y2):
        return True
    else:
        return False

```

```python
are_rectangles_intersecting(((0, 0), (5, 5)), ((1, 1), (2, 2)))  
```

```python
# Are rectangles (with sides parallel to axes) intersecting? # Works for all cases

def are_rectangles_intersecting(rect1, rect2):
    # UNPACK the tuples (variables on the left)
    (r1x1, r1y1), (r1x2, r1y2) = rect1
    (r2x1, r2y1), (r2x2, r2y2) = rect2

    # Check for overlap on both axes (The "No Gap" rule)
    x_intersect = max(min(r1x1, r1x2), min(r2x1, r2x2)) <= min(max(r1x1, r1x2), max(r2x1, r2x2))
    y_intersect = max(min(r1y1, r1y2), min(r2y1, r2y2)) <= min(max(r1y1, r1y2), max(r2y1, r2y2))

    return x_intersect and y_intersect
```

```python
are_rectangles_intersecting(((0, 0), (5, 5)), ((1, 1), (2, 2))) 
```

```python
# Is point on line segment in 2D
def is_point_on_segment(A, B, P):
    (ax, ay), (bx, by), (px, py) = A, B, P
    if (min(ax, bx) <= px <= max(ax, bx) and min(ay, by) <= py <= max(ay, by)):
        return True
    else:
        return False
```

```python
A, B, P = (0, 0), (10, 10), (5, 5)
print(is_point_on_segment(A, B, P))
```

```python
# Device impurity formula for a set with only two classes given the counts of each class.

def my_impurity(c1, c2):
    total = c1 + c2
    if total == 0:
        return 0.0
    return 1 - (abs(c1 - c2) / total)
```

```python
my_impurity(9, 1) 
```

```python
# Can you design a formula that gives values between 0 and 1 and changes smoothly with the ratio of the classes?

def my_smooth_impurity(c1, c2):
     total = c1 + c2
     if total == 0:
        return 0.0
     return (4 * c1 * c2) / (total ** 2)
```

```python
my_smooth_impurity(9, 1) 
```
